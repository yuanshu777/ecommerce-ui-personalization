# Project Overview: UI A/B Experiment and Conversion Impact

## 1. Business Question

The core business question of this project is:

> **Does replacing the default UI with alternative (non-default) UI designs causally improve high-intent user conversion, and if so, for which types of users is the effect strongest?**

This question is directly tied to product decision-making, specifically whether UI changes should be rolled out broadly, selectively, or further optimized.

---

## 2. Data & Unit of Analysis

* **Unit of analysis:** Session-level
* **Experiment type:** Randomized A/B experiment
* **Treatment:** Exposure to non-default UI variants
* **Control:** Default UI

Each session is independently assigned to a UI variant, enabling causal inference under standard A/B testing assumptions.

---

## 3. Key Metrics

### Primary Metric

* **High-intent conversion rate** (binary, session-level)

### Secondary / Diagnostic Metrics

* Bounce rate
* User engagement (event count, session duration)

Secondary metrics are used to understand behavioral mechanisms and to assess heterogeneity, not to replace the primary decision metric.

---

## 4. Experiment Design & Statistical Framework

### Estimand

> **Difference in session-level high-intent conversion rate between default UI and non-default UI variants.**

### Statistical Methods

* Two-proportion z-tests
* 95% confidence intervals for effect size estimation
* Multiple hypothesis testing correction using **False Discovery Rate (FDR)**
* Sensitivity / robustness analyses under different engagement and filtering thresholds

This framework prioritizes effect size, uncertainty, and decision reliability over single p-value reporting.

---

## 5. Variant-Level and Heterogeneity Analysis

Beyond average treatment effects, the analysis examines:

* Performance differences across individual UI variants
* Stability of lift estimates across filters
* **Heterogeneous treatment effects** stratified by user engagement levels

This allows identification of both globally strong variants and segments where UI changes are particularly effective.

---

## 6. Modeling for Validation (Non-Causal)

Predictive models are used **to support and validate experimental findings**, not for causal inference.

* **Logistic Regression:** interpretability and directionality of effects
* **XGBoost:** non-linear interactions and performance benchmarking

### Evaluation Metrics

* ROC-AUC
* PR-AUC
* Log Loss
* Brier Score

Model results consistently support the experimental conclusion that UI design has an independent effect beyond engagement proxies.

---

## 7. Key Findings

* Non-default UI variants **significantly improve high-intent conversion** relative to the default UI
* Observed absolute lift ranges approximately **2.5–8.5 percentage points**, depending on variant and segment
* Results remain statistically significant after **FDR adjustment (p < 0.05)**
* Treatment effects are **stronger among high-engagement users**
* Conclusions are robust across multiple sensitivity analyses

---

## 8. Product Decisions & Recommendations

Based on the combined experimental and modeling evidence:

1. **Roll out the most stable, high-performing non-default UI variants**
2. Prioritize **UI optimization for high-engagement users**, where marginal impact is highest
3. Consider next steps such as **personalized UI or uplift modeling** to further target UI exposure

---

## 9. Scope & Explicit Non-Goals

To maintain clarity and reliability, the following were intentionally out of scope:

* Sequential testing or adaptive experimentation
* CUPED or variance reduction techniques
* Causal machine learning or uplift models

These methods are viable extensions but were deprioritized in favor of robustness, interpretability, and clear decision support.

---

## 10. Summary

This project demonstrates an end-to-end, decision-oriented data science workflow that combines **causal experimentation**, **robust statistical validation**, and **supporting predictive modeling** to inform real product decisions around UI design and conversion optimization.
